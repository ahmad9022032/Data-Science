# Task 1 "Bar Plot " : <a id="4"></a> 

```python
df_can.set_index('Country', inplace=True)

# Let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))
df_continents.head()
```

```python
df_can.sort_values(['Total'],ascending = False, axis=0, inplace =True)
years = list(map(str, range(1980, 2014)))
```

```python
df_continents = df_continents[years].transpose()
df_continents
```

### Continent wise bar plotting: 

```python
df_continents.plot(kind = 'bar', figsize = (10,6))
plt.title('continent wise imigrants to Canada from 1980 to 2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```

#  Task 2 "Histogram " : <a id="4"></a> 

```python
df_can['2010'].head()
```

```python
# np.histogram returns 2 values
count, bin_edges = np.histogram(df_can['2010'])

print('count >> ',count) # frequency count
print('bin edges >> ',bin_edges) # bin ranges, default = 10 bins
```

```python
df_can['2010'].plot(kind='hist', figsize=(8, 5))

# add a title to the histogram
plt.title('Histogram of Immigration  in 2010')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')

plt.show()
```

#  Task 3 "Box Plot " : <a id="4"></a> 

```python
# Group by continent and sum the immigration data
df_continents = df_can.groupby('Continent', axis=0).sum()

# Printing the type of df_can.groupby('Continent', axis=0)
print(type(df_can.groupby('Continent', axis=0)))

# Display the head of df_continents
print(df_continents.head())
```

```python
# code to filter the dataset for the specified year (2000) and extract immigration data for each region
year = '2000'
regions = df_continents.index
region_immigration_data = {region: df_continents.loc[region, year] for region in regions}
```

```python
# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(region_immigration_data.values(), patch_artist=True, showfliers=True)

# Customize the box plot
plt.xlabel('Region')
plt.ylabel('Immigration')
plt.title(f'Immigration Distribution Across Regions in {year}')

# Show plot
plt.grid(True)
plt.show()
```

#  Task 4 "Scatter Plot " : <a id="4"></a> 

```python
# create df_countries by selecting immigration data for Denmark, Norway, and Sweden for the years 2000 to 2010
df_countries = df_can.loc[['Denmark', 'Norway', 'Sweden'], '2000':'2010'].transpose()

# create df_total by summing across three countries for each year
df_total = pd.DataFrame(df_countries.sum(axis=1))

# reset index in place
df_total.reset_index(inplace=True)

# rename columns
df_total.columns = ['year', 'total']

# change column year from string to int to create scatter plot
df_total['year'] = df_total['year'].astype(int)

# show resulting dataframe
print(df_total.head())
```

```python
# generate scatter plot
df_total.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

 # add title and label to axes
plt.title('Immigration from Denmark, Norway, and Sweden to Canada from 2000 - 2010')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

 # show plot
plt.show()
```

#  Task 5 "Combined Visualization Task " : <a id="4"></a> 

```python
# Calculate total immigration for each continent for the desired range of years
df_continents_years = df_continents.loc[:, '1980':'2013']
```

```python
# Plot multi-series line plot
plt.figure(figsize=(10, 8))
for continent in df_continents_years.index:
    plt.plot(df_continents_years.columns, df_continents_years.loc[continent], label=continent)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Immigration')
plt.title('Immigration Trends Across Continents (1980-2013)')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
